```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

print("All libraries imported successfully!")

def load_sample_images(folder_path, num_samples=5):
    images = []
    for root, _, files in os.walk(folder_path):
        for file_name in files[:num_samples]:
            img_path = os.path.join(root, file_name)
            try:
                image = Image.open(img_path)
                images.append(image)
            except Exception as e:
                print(f"Could not open {img_path}: {e}")
            if len(images) >= num_samples:
                break
    return images

# Example usage for ETL1 dataset
etl1_samples = load_sample_images('data/raw/ETL1', num_samples=5)

def display_images(images, title):
    plt.figure(figsize=(10, 5))
    for i, img in enumerate(images):
        plt.subplot(1, len(images), i+1)
        plt.imshow(img, cmap='gray')
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

display_images(etl1_samples, "Sample Katakana Characters from ETL1 Dataset")

# # Example: Load labels (assuming labels are stored in a CSV file)
# labels = pd.read_csv('data/raw/katakana_labels.csv')  # Adjust path as necessary

# # Plot distribution
# plt.figure(figsize=(12, 6))
# labels['character'].value_counts().plot(kind='bar')
# plt.title('Distribution of Katakana Characters in Dataset')
# plt.xlabel('Character')
# plt.ylabel('Frequency')
# plt.show()

def check_image_sizes(folder_path):
    sizes = []
    for root, _, files in os.walk(folder_path):
        for file_name in files:
            img_path = os.path.join(root, file_name)
            try:
                with Image.open(img_path) as img:
                    sizes.append(img.size)
            except Exception as e:
                print(f"Could not open {img_path}: {e}")
    unique_sizes = set(sizes)
    print("Unique image sizes:", unique_sizes)

check_image_sizes('data/raw/ETL1')
```
